```python
import numpy as np
print(np.__version__)
```

```python
arr = np.ones((2,3))
print(arr)
```

```python
arr = np.zeros((2,3))
print(arr)
```

```python
arr = np.ones((2,3))
print(arr)
print(arr.ndim)
print(arr.shape)
print(arr.dtype)
print(type(arr))
print(arr.size)
print(arr.itemsize)
print(arr.nbytes)
```

```python
arr = np.full((4,5),8)
print(arr)
```

```python
arr = np.arange(1,10,2)
print(arr)
arr = np.random.random((4,5))
print(arr)
arr = np.random.randint(1,5,(3,4))
print(arr)
```

```python
arr = np.eye(5)
print(arr)
arr = np.linspace(1,5,10)
print(arr)
```

```python
print(arr.sum())
print(np.sum(arr))
print(arr.mean())
print(np.mean(arr))
print(np.median(arr))
print(arr.max())
print(arr.min())
print(arr.std())
print(arr.var())
```
