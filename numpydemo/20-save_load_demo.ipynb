```python
import numpy as np
import matplotlib.pyplot as plt

try:
    logo = np.load("numpy_logo.npy")
except FileNotFoundError:
    print("File not found")

print("dtype:", logo.dtype)
print("min/max:", logo.min(), logo.max())
print("shape:", logo.shape)

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(logo)
plt.title("Numpy logo")

dark_logo = logo.copy()
dark_logo[..., :3] = 255 - dark_logo[..., :3]

#plt.figure(figsize=(7,5))
plt.subplot(122)
plt.imshow(dark_logo)
plt.axis("off")
plt.title("Numpy logo Dark Mode")

```
