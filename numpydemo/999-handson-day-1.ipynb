### Array creation

```python
array_0 = np.array(10)
print(array_0)
print(type(array_0))
print(array_0.ndim)
print(array_0.shape)
print(array_0.dtype)
print(array_0.size)
print(array_0.itemsize)
print(array_0.nbytes)
```

```python
array_0 = np.arange(1,10,1)
print(array_0)
print(type(array_0))
print(array_0.ndim)
print(array_0.shape)
print(array_0.dtype)
print(array_0.size)
print(array_0.itemsize)
print(array_0.nbytes)
```

```python
array_1d = np.array([35,2,6,78,235,7,4])
print(np.sort(array_1d))
```

```python
array_2d = np.array([[2,1],[3,2],[1,3]])
print(np.sort(array_2d))
print(np.sort(array_2d,axis=0))
print(np.sort(array_2d,axis=1))
```

```python
array_1d = np.array([10,3,5,7,1,3])
print(array_1d[array_1d > 3])
print(array_1d > 3)
bool_array = array_1d > 3
print(~bool_array)
print(array_1d[~bool_array])
```
