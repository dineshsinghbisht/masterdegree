### Array creation

```python
array_0 = np.array(10)
print(array_0)
print(type(array_0))
print(array_0.ndim)
print(array_0.shape)
print(array_0.dtype)
print(array_0.size)
print(array_0.itemsize)
print(array_0.nbytes)
```

```python
array_0 = np.arange(1,10,1)
print(array_0)
print(type(array_0))
print(array_0.ndim)
print(array_0.shape)
print(array_0.dtype)
print(array_0.size)
print(array_0.itemsize)
print(array_0.nbytes)
```

```python
array_2 = np.random.randint(1,10,(2,3))
print(array_2)
print(type(array_2))
print(array_2.ndim)
print(array_2.shape)
print(array_2.dtype)
print(array_2.size)
print(array_2.itemsize)
print(array_2.nbytes)
```
