```python
import numpy as np # pip install numpy
print(np.__version__) # to get version
```

### Creating 1D Array and properties

```python
arr1d = np.array([1,14,67])
arr1d = np.arange(1,10) # start,end(exclusive),step
arr1d = np.random.randint(1,10,(5))
print("Array value ", arr1d)
print("Type is", type(arr1d))
print("Shape is", arr1d.shape, np.shape(arr1d))
print("Size is", arr1d.size, np.size(arr1d))
print("Length is", len(arr1d))
print("Data type is", arr1d.dtype)
print("Itemsize is", arr1d.itemsize)
print("Size is", arr1d.nbytes)
```

### creating arrays with different ways

```python
print("Example for np.zeroes((4,5))\n", np.zeros((4,5)))
print(np.ones((4,5)))
print(np.full((3,4),100))
print(np.eye(5))
print(np.arange(1,10,2))
print(np.linspace(1,10,2))
print(np.random.random((4,5)))
print(np.random.rand(4,5))
print(np.random.randint(1,10,(3,4)))
print(np.empty((4,5)))
```

### Reshape, Flatten, Ravel, Transpose

```python
arr = np.arange(1,10)
print("Initial array is", arr)
print("reshape array is", arr.reshape(3,-1))
print("reshape array is", arr.reshape(3,3))
print("reshape array is", arr.reshape(9))
print("reshape array is", np.reshape(arr,(3,3)))
print("Flatten is", arr.flatten())
print("Ravel is", arr.ravel())
print("Original array is", arr)

arr2d = np.random.randint(1,10, (2,4))
print("\nOriginal array is", arr2d)
print("Transpose 2d array is", arr2d.T)
```

### Indexing and Slicing in 1D Arrays

```python
arr1d = np.random.randint(1,100,10)
print("Array is", arr1d)
print("Value of arr1d[3] is", arr1d[3])
print("Value of arr1d[-3] is", arr1d[-3])
print("Value of arr1d[3:] is", arr1d[3:])
print("Value of arr1d[::2] is", arr1d[::2])
print("Value of arr1d[::-1] is", arr1d[::-1])
print("Original Array is", arr1d)
```

### Indexing and Slicing for 2d Arrays

```python
arr2d = np.random.randint(1,100,(4,5))
print("Value of array is\n", arr2d)
print("value of arr2d[0] is \n", arr2d[0])
print("value of arr2d[0:] is \n", arr2d[0:])
print("value of arr2d[:,1] is \n", arr2d[:,1])
print("value of arr2d[:,1:] is \n", arr2d[:,1:])
print(arr2d[1:3,1:3])
```

### Sorting 1d array

```python
arr1d = np.array([13,43,4,867,55])
print("Original array is", arr1d)
print("Sorting is", np.sort(arr1d))
print("Original array is", arr1d)
print("Sorting is", arr1d.sort()) # it sorts the original array
print("Original array is", arr1d)
```

### Sort 2d Arrays

```python
arr2d = np.array([[3,1],[2,3],[1,2]])
print("Original array is\n", arr2d)
print("Sort array column wise\n", np.sort(arr2d,axis=0))
print("Sort array row wise\n", np.sort(arr2d,axis=1))
print("Sort array default\n", np.sort(arr2d))
```

### Boolean filter

```python
arr1d = np.array([1,4,6,7,8,9,10,6])
print(arr1d % 2 == 0)
print(arr1d[arr1d % 2 == 0])
print(arr1d > 5)
print(arr1d[(arr1d > 5)])
print((arr1d > 4) & (arr1d < 10))
print(arr1d[(arr1d > 4) & (arr1d < 10)])
print(~(arr1d > 4) & (arr1d < 10))
```

### where filter

```python
arr1d = np.array([1,4,6,7,8,9,10,6])
print(np.where(arr1d > 6))
print(arr1d[np.where(arr1d > 6)])
print(np.where(arr1d > 6, True, False))
```

### Append Vs concatenate Vs insert

```python
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
# print(np.append(arr1,arr2))
# print(np.concatenate((arr1,arr2))) # arrays passed as tuple

arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr1)
print(arr2)
print(np.append(arr1,arr2))
print("Output of append\n",np.append(arr1,arr2,axis=0))
print("Output of concatenate\n",np.concatenate((arr1,arr2),axis=0))
print(np.append(arr1,arr2,axis=1))
print(np.concatenate((arr1,arr2),axis=1))

arr = np.array([[1,2],[3,4]])
print(np.insert(arr,0,[5,6],axis=0))
print(np.insert(arr,1,[5,6],axis=0))
print(np.insert(arr,2,[5,6],axis=0))
print(np.insert(arr,0,[5,6],axis=1)) # NumPy reshapes it internally to (1, 2)
print(np.insert(arr,2,[5,6],axis=1))
print(np.insert(arr,2,[5,6],axis=None))
print(np.concatenate((arr,[[5,6]]),axis=0))
print(np.concatenate((arr,[[5],[6]]),axis=1))
```

### vstack and hstack

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr1)
print(arr2)
print(np.hstack((arr1,arr2)))
print(np.vstack((arr1,arr2)))
```
