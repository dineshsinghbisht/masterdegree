```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D"]
sales_2024 = [20, 35, 30, 25]
sales_2025 = [25, 32, 34, 28]

plt.bar(categories, sales_2024, label="2024")
plt.bar(categories, sales_2025, bottom=sales_2024, label="2025")

plt.legend(title="Year")
plt.show()
```

```python
labels = ["AWS", "Azure", "GCP", "Other"]
sizes = [40, 30, 20, 10]

plt.pie(
    sizes, 
    labels=labels, 
    autopct="%1.1f%%")
plt.title("Cloud Market Share")
plt.show()
```

```python
colors = ["skyblue", "orange", "green", "gray"]
sizes = [40, 30, 20, 10]
labels = ["AWS", "Azure", "GCP", "Other"]

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Cloud Market Share")
plt.axis("equal")  # Makes it a circle
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.normal(loc=50, scale=10, size=1000)
print(data)
print(len(data))

plt.hist(data, bins=20)
plt.title("Histogram of Test Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

plt.scatter(x, y)
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Hours vs Score")
plt.show()

plt.scatter(
    x, y,
    color="blue",
    marker="+",
    s=100,        # size of points
    alpha=0.7     # transparency
)
plt.show()


x1 = [1, 2, 3, 4]
y1 = [10, 15, 20, 25]

x2 = [1, 2, 3, 4]
y2 = [12, 18, 22, 28]

plt.scatter(x1, y1, label="Group A")
plt.scatter(x2, y2, label="Group B")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(121)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Left Plot")

plt.subplot(122)
plt.plot([1, 2, 3], [3, 2, 1])
plt.title("Right Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt 
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].plot([1, 2, 3], [1, 4, 9], label="Model A")
axs[1].plot([1, 2, 3], [3, 2, 1], label="Model B")

fig.suptitle("Model Performance")
fig.legend(loc="upper left")
fig.tight_layout(rect=[0, 0, 1, 0.9])

plt.show()
```
