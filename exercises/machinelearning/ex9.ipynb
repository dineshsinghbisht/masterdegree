```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression

# Load sensor_readings_24.csv dataset into a pandas DataFrame
df = pd.read_csv("sensor_readings_24.csv")

# # Get basic information about the data
# print(df.info())
# print(df.head(5))
# print(df.isna().sum())  # Check missing values, handle them (drop or impute) if exists
# # No missing values found in the selected features and target columns; imputation is not required.

# Extract all the columns except the last as features (X)
X = df.iloc[:, :-1]

# Scale all the numeric features using StandardScaler
scaler_x = StandardScaler()
X_scaled = scaler_x.fit_transform(X)

# Extract the last column as target (y) and encode class labels into numeric values using LabelEncoder
enc = LabelEncoder()
y = enc.fit_transform(df.iloc[:, -1])

# Split the data into training (75%) and test (25%) datasets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25)

# Fit Logistic Regression model on the training dataset
log_reg_model = LogisticRegression(solver="newton-cg") # 'multi_class' option was deprecated in version 1.5 and will be removed in 1.8
log_reg_model.fit(X_train,y_train)

# Generate predictions for training and test datasets
log_reg_model_prediction_train = log_reg_model.predict(X_train)
log_reg_model_prediction_test = log_reg_model.predict(X_test)

# Evaluate and display model accuracy with Accuracy Score for both training and test datasets
print(f"Accuracy Score for training dataset using Logistic Regression model: {accuracy_score(y_train, log_reg_model_prediction_train):.2f}")
print(f"Accuracy Score for test dataset using Logistic Regression model: {accuracy_score(y_test, log_reg_model_prediction_test):.2f}")
```
