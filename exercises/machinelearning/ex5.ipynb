```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

input_data = {'X': [1.00, 2.00, 3.00, 4.00, 5.00],
        'Y': [1.00, 2.00, 1.30, 3.75, 2.25]}

# Create dataframe using input_data
df = pd.DataFrame(input_data)
print(df.info())

# Separate features and response columns
X = df[["X"]]   # feature
y = df["Y"]     # response

# Check the types of feature and reponse
print(type(X))
print(type(y))

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the value when X = 6
X_pred = pd.DataFrame({"X": [6]})
y_pred = model.predict(X_pred)[0]
print("Prediction for X = 6 is", y_pred)

# Plot scatter 
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Input Data")
plt.plot(X, model.predict(X), label="Regression Line")
plt.scatter(X_pred, y_pred, label="Prediction for X=6)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
