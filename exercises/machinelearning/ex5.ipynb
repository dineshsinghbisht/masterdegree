```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Using input data create pandas DataFrame
df = pd.DataFrame({"X": [1.00, 2.00, 3.00, 4.00, 5.00],
                   "Y": [1.00, 2.00, 1.30, 3.75, 2.25]})

# # Get basic information about data (feature and target) specifically datatypes
# print(df.info())

# Separate feature (X) and target (y) from the DataFrame
X = df[["X"]]
y = df["Y"]

# Fit Linear Regression model on the data
linreg_model = LinearRegression()
linreg_model.fit(X, y)


# Prepare input data for prediction
X_6 = pd.DataFrame({"X": [6]}) 

# Generate prediction when X = 6
y_pred = linreg_model.predict(X_6)
print(f"Prediction when X = 6 is {y_pred[0]:.2f}")

# Plot actual data, prediction at X=6, and regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Actuals")
plt.scatter(X_6, y_pred, label="Prediction when X=6")
plt.plot(X, linreg_model.predict(X), label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
