```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create dataframe using input data
df = pd.DataFrame({'X': [1.00, 2.00, 3.00, 4.00, 5.00],
        'Y': [1.00, 2.00, 1.30, 3.75, 2.25]})

# # check information about dataframe specifically datatypes
# print(df.info())

# Separate feature and response from dataframe
X = df[["X"]]   # feature
y = df["Y"]     # response

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the value when X = 6
X_pred = pd.DataFrame({"X": [6]}) # input data
y_pred = model.predict(X_pred)
print(f"Prediction for X = 6 is {y_pred[0]:.2f}")

# Plot scatters and line plots for given data and prediction
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Actual")
plt.plot(X, model.predict(X), label="Regression Line")
plt.scatter(X_pred, y_pred, label="Prediction when X=6")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create dataframe using input_data
df = pd.DataFrame({'X': [1.00, 2.00, 3.00, 4.00, 5.00],
        'Y': [1.00, 2.00, 1.30, 3.75, 2.25]})

# check information about dataframe
print(df.info())

# Separate feature and response from dataframe
X = df[["X"]]   # feature
y = df["Y"]     # response

# Create and train the linear regression model using input data
model = LinearRegression()
model.fit(X, y)

# Predict the value when X = 6
X_pred = pd.DataFrame({"X": [6]}) # input data
# y_pred = model.predict(X_pred)[0]
y_pred = model.predict(X_pred)
print(f"Prediction for X = 6 is {y_pred[0]:.2f}")

# Plot scatters and line plots 
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Actual")
plt.plot(X, model.predict(X), label="Regression Line")
plt.scatter(X_pred, y_pred, label="Prediction for X=6")
plt.xlabel("X") 
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
