```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create Pandas DataFrame using input data
df = pd.DataFrame({"X": [1.00, 2.00, 3.00, 4.00, 5.00],
        "Y": [1.00, 2.00, 1.30, 3.75, 2.25]})

# # Get basic information about data (feature and response) specifically datatypes
# print(df.info())

# Separate feature and response from DataFrame
X = df[["X"]]   # feature / Input
y = df["Y"]     # response / Output

# Fit Linear Regression Model to training dataset
linreg_model = LinearRegression()
linreg_model.fit(X, y)

# Predict the value when X = 6
X_6 = pd.DataFrame({"X": [6]}) # prepare input data for prediction
y_pred = linreg_model.predict(X_6)
print(f"Prediction for X = 6 is {y_pred[0]:.2f}")

# Plot scatters and line plots for given data and prediction
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Actual")
plt.scatter(X_6, y_pred, label="Prediction when X=6")
plt.plot(X, linreg_model.predict(X), label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
