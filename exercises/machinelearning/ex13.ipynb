```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import tensorflow as tf

# Load heart.csv dataset into a pandas DataFrame 
df = pd.read_csv("heart.csv")

# # Get basic information about the data
# print(df.info())
# print(df.head(5))
# print(df.isna().sum())
# # No missing values found in the data; imputation is not required.

# Extract all the columns except the last as features (X)
X = df.iloc[:, :-1]

# Scale all the features using StandardScaler
scaler_x = StandardScaler()
X_scaled = scaler_x.fit_transform(X)

# Extract the last column as target (y) and converts into one-hot encoded dummy variables
y = pd.get_dummies(df.iloc[:, -1])

# Split the data into training (75%) and test (25%) datasets
# X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, stratify=y)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, stratify=y, random_state=42)

# Create, compile and fit a Neural Network model on the training dataset
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(50, activation='relu',),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(30, activation='relu',),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(y_train.shape[1], activation='softmax')
])

model.compile(
    tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(
    X_train, y_train, 
    validation_data=(X_test, y_test), 
    epochs=50, batch_size=15
)

model.fit(X_train,y_train)

# Generate predictions for training and test datasets
nn_model_prediction_train = np.argmax(model.predict(X_train), axis=1)
nn_model_prediction_test = np.argmax(model.predict(X_test), axis=1)

# Evaluate and display model accuracy with Accuracy Score for training and test datasets
print(f"\nAccuracy Score for training dataset using Neural Network model: {accuracy_score(np.argmax(y_train, axis=1), nn_model_prediction_train):.2f}")
print(f"Accuracy Score for test dataset using Neural Network model: {accuracy_score(np.argmax(y_test, axis=1), nn_model_prediction_test):.2f}\n")
```
