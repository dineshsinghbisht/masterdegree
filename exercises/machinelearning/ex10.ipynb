```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC

# Load csv containing Robot Sensor Reading data into Pandas DataFrame
df = pd.read_csv("sensor_readings_24.csv")

# # Get basic information about data
# print(df.info())
# print(df.head(5))

# Extract all the columns except the last as features (X)
X = df.iloc[:, :-1]

# Standardize all numeric features using StandardScaler
scaler_x = StandardScaler()
X_scaled = scaler_x.fit_transform(X)

# Extract the last column as target (y) and encode class labels into numeric values using LabelEncoder
enc = LabelEncoder()
y = enc.fit_transform(df.iloc[:, -1])

# Split data into training (75%) and test (25%) datasets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25)

# Fit Support Vector Machine model on the training dataset
svm_model = SVC()
svm_model.fit(X_train,y_train)

# Generate predictions for training and test datasets
svm_model_prediction_train = svm_model.predict(X_train)
svm_model_prediction_test = svm_model.predict(X_test)

# Evaluate and display model accuracy with Accuracy Score for both training and test datasets
print(f"Accuracy Score in training dataset using Support Vector Machine model: {accuracy_score(y_train, svm_model_prediction_train):.2f}")
print(f"Accuracy Score in test dataset using Support Vector Machine model: {accuracy_score(y_test, svm_model_prediction_test):.2f}\n")

# Convert the test predictions back to original class labels using inverse transform
svm_model_prediction_test_classes = enc.inverse_transform(svm_model_prediction_test)

# Create an empty DataFrame to store validation results
df_validation = pd.DataFrame()

# Add predicted class labels of the test dataset
df_validation["Prediction"] = svm_model_prediction_test_classes

# Add actual class labels of the test dataset
df_validation["Actual Command"] = enc.inverse_transform(y_test)

# Print randomly selected 20 rows
df_validation = df_validation.sample(20)
print("Predicted and Actual Control Command for 20 random rows from the test data")
print(df_validation.to_string(index=False))
```
