```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Load the dataset from a csv file into a Pandas DataFrame 
df = pd.read_csv("covtype.csv")

# # Get basic information about data
# print(df.info())
# print(df.columns)
# print(df.head())
# print(df.isna().sum())

# Prepare features (X) and target (y)
X = df.iloc[:, :-1]     # extract all the columns except last
y = df.iloc[:, -1]      # extract last column only

# Repeat the train-test split and model training 100 times for repeated validation - Decision Tree Classifier
dtc_accuracies_train = []
dtc_accuracies_test = []
for i in range(0, 100):
    # Split data into train (70%) and test (30%) sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)

    # Create and train a Decision Tree classifier with maximum depth of 3
    dtc_model = DecisionTreeClassifier(max_depth=3)
    dtc_model.fit(X_train, y_train)

    # Generate predictions for the train and test dataset
    dtc_model_prediction_train = dtc_model.predict(X_train)
    dtc_model_prediction_test = dtc_model.predict(X_test)

    # Evaluate and display model accuracy with Accuracy Score for train and test datasets
    print(f"Accuracy Score in train dataset using Decision Tree Classifier Model: {accuracy_score(y_train, dtc_model_prediction_train):.2f}")
    print(f"Accuracy Score in test dataset using Decision Tree Classifier Model: {accuracy_score(y_test, dtc_model_prediction_test):.2f}")

    dtc_accuracies_train.append(accuracy_score(y_train, dtc_model_prediction_train))
    dtc_accuracies_test.append(accuracy_score(y_test, dtc_model_prediction_test))

print(f"\nMean train data accuracy with Decision Tree Classifier Model {np.mean(dtc_accuracies_train):.2f}")
print(f"Mean test data accuracy with Decision Tree Classifier Model {np.mean(dtc_accuracies_test):.2f}")

# Plot Decision Tree Graph
plt.figure(figsize=(10, 6))
plot_tree(dtc_model, feature_names=df.iloc[:, :-1].columns, filled=True)
plt.show()


# Repeat the train-test split and model training 100 times for repeated validation - Random Forest Classifier
rfc_accuracies_train = []
rfc_accuracies_test = []
for i in range(0, 100):
    # Split data into train (70%) and test (30%) sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)

    # Create and train a Random Forest classifier with maximum depth of 3
    rfc_model = RandomForestClassifier(max_depth=3)
    rfc_model.fit(X_train, y_train)

    # Generate predictions for the train and test dataset
    rfc_model_prediction_train = rfc_model.predict(X_train)
    rfc_model_prediction_test = rfc_model.predict(X_test)

    # Evaluate and display model accuracy with Accuracy Score for train and test datasets
    print(f"Accuracy Score in train dataset using Random Forest Classifier Model: {accuracy_score(y_train, rfc_model_prediction_train):.2f}")
    print(f"Accuracy Score in test dataset using Random Forest Classifier Model: {accuracy_score(y_test, rfc_model_prediction_test):.2f}")

    rfc_accuracies_train.append(accuracy_score(y_train, rfc_model_prediction_train))
    rfc_accuracies_test.append(accuracy_score(y_test, rfc_model_prediction_test))

print(f"\nMean train data accuracy with Random Forest Classier Model {np.mean(rfc_accuracies_train):.2f}")
print(f"Mean test data accuracy with Random Forest Classier Model {np.mean(rfc_accuracies_test):.2f}")
```
