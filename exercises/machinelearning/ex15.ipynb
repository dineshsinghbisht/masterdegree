```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Load covtype.csv dataset into a pandas DataFrame
df = pd.read_csv("covtype.csv")

# # Get basic information about the data
# print(df.info())            # display information about dataframe
# print(df.head(5))           # display first 5 rows
# print(df.isna().sum())      # Check missing values, handle them (drop or impute) if exists
# # No missing values found in the data; imputation is not required.

# Prepare features (X) and target (y)
X = df.iloc[:, :-1]     # extract all the columns except last
y = df.iloc[:, -1]      # extract last column only

# Repeat the train-test split and model training 100 times for repeated validation - Decision Tree Classifier
dtc_accuracies_train = []
dtc_accuracies_test = []
for i in range(0, 100):
    # Split data into train (70%) and test (30%) datasets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)

    # Fit Decision Tree Classifier model on the training dataset
    dtc_model = DecisionTreeClassifier(max_depth=3)
    dtc_model.fit(X_train, y_train)

    # Generate predictions for training and test dataset
    dtc_model_prediction_train = dtc_model.predict(X_train)
    dtc_model_prediction_test = dtc_model.predict(X_test)

    # Evaluate and display model accuracy with Accuracy Score for training and test datasets
    print(f"Accuracy Score for training dataset using Decision Tree Classifier model: {accuracy_score(y_train, dtc_model_prediction_train):.2f}")
    print(f"Accuracy Score for test dataset using Decision Tree Classifier model: {accuracy_score(y_test, dtc_model_prediction_test):.2f}")

    dtc_accuracies_train.append(accuracy_score(y_train, dtc_model_prediction_train))
    dtc_accuracies_test.append(accuracy_score(y_test, dtc_model_prediction_test))

print(f"\nMean training dataset accuracy with Decision Tree Classifier model {np.mean(dtc_accuracies_train):.2f}")
print(f"Mean test dataset accuracy with Decision Tree Classifier model {np.mean(dtc_accuracies_test):.2f}")

# Plot Decision Tree Graph
plt.figure(figsize=(20, 12), dpi=150)
plot_tree(dtc_model, feature_names=df.iloc[:, :-1].columns, filled=True)
plt.show()

# Repeat the train-test split and model training 100 times for repeated validation - Random Forest Classifier
rfc_accuracies_train = []
rfc_accuracies_test = []
for i in range(0, 100):
    # Split data into train (70%) and test (30%) datasets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)

    # Fit Random Forest Classifier model on the training dataset
    rfc_model = RandomForestClassifier(max_depth=3)
    rfc_model.fit(X_train, y_train)

    # Generate predictions for training and test dataset
    rfc_model_prediction_train = rfc_model.predict(X_train)
    rfc_model_prediction_test = rfc_model.predict(X_test)

    # Evaluate and display model accuracy with Accuracy Score for training and test datasets
    print(f"Accuracy Score for training dataset using Random Forest Classifier model: {accuracy_score(y_train, rfc_model_prediction_train):.2f}")
    print(f"Accuracy Score for test dataset using Random Forest Classifier model: {accuracy_score(y_test, rfc_model_prediction_test):.2f}")

    rfc_accuracies_train.append(accuracy_score(y_train, rfc_model_prediction_train))
    rfc_accuracies_test.append(accuracy_score(y_test, rfc_model_prediction_test))

print(f"\nMean training dataset accuracy with Random Forest Classier model {np.mean(rfc_accuracies_train):.2f}")
print(f"Mean test dataset accuracy with Random Forest Classier model {np.mean(rfc_accuracies_test):.2f}")
```
