### Problem 1. Wines


[Here](https://student.labranet.jamk.fi/~varpha/data_analytics/exrc06p01_wine.csv) is some data on Portuguese wines. 

Drop rows with missing values.

Use logistic regression to predict the type (white/red) from the other fields.

Split train/test set 70/30 %. Print the score and the confusion matrix.


````python
import pandas as pd

df = pd.read_csv('data/exrc06p01_wine.csv')
df=df.dropna()
df.head()
````

````python
df.info()
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler


# the logistic regression model
classifier = LogisticRegression()

X = df.drop('type', axis=1)
y = df['type']

# scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

````python

from sklearn.model_selection import train_test_split

# splitting to train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3,  random_state=42)

classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)


print('score (scaled fit):', classifier.score(X_test, y_test))

print(y_pred)
````

````python

from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
````

### Problem 2. Voices
[Here](https://student.labranet.jamk.fi/~varpha/data_analytics/exrc06p02_voice.csv) is some data on human voices ([column info](https://student.labranet.jamk.fi/~varpha/data_analytics/exrc06p02_voice.txt)).
 
Predict the label from the other fields using a support vector machine.

Split train/test set 70/30 %.

Print the score and the confusion matrix.


````python
import pandas as pd

df = pd.read_csv('data/exrc06p02_voice.csv')
df=df.dropna()
df.head()
````

````python
df.info()
````

````python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC


X = df.drop('label', axis=1)
y = df['label']

my_classifier = SVC()


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

    
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3,  random_state=42)

my_classifier.fit(X_train, y_train)
y_pred = my_classifier.predict(X_test)

print('score (scaled fit):', my_classifier.score(X_test, y_test))
print(y_pred[:50])
````

````python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
````

````python
# The same with pipeline and without random state

import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

my_classifier = make_pipeline(
    StandardScaler(),
    SVC()
)
# my_classifier = SVC() # classify with support vector
# override the pipeline above to ignore the scaling part...

myscaler = StandardScaler()
# ... and do the scaling separately

df = pd.read_csv(
    'data/exrc06p02_voice.csv'
)

df=df.dropna()

X = df.drop('label', axis=1)
y = df['label']

myscaler.fit(X)
# same as StandardScaler().fit()
# myscaler now contains the info about how
# the columns of X need to be scaled to normality

X_scaled = pd.DataFrame(
    myscaler.transform(X), # the actual scaling happens here
    columns=X.columns
)
# a mere myscaler.transform(X) would produce an array
# but that's minor... made a dataframe in order to compare:

# print(X)
# print(X_scaled)
# comment out the above to see the unscaled and scaled data

# first fit without scaling
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3)
my_classifier.fit(X_train, y_train)
y_pred = my_classifier.predict(X_test)

print(
    'score (unscaled fit):',
    my_classifier.score(X_test, y_test)
)

# then fit with scaling
# note: X_scaled used below
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y,test_size=0.3)
my_classifier.fit(X_train, y_train)
y_pred = my_classifier.predict(X_test)

print(
    'score (scaled fit):',
    my_classifier.score(X_test, y_test)
)
````

### Problem 3. NBA
[Here](https://student.labranet.jamk.fi/~varpha/data_analytics/exrc06p03_nba.csv) is some data on NBA basketball players in their first season ([column info](https://student.labranet.jamk.fi/~varpha/data_analytics/exrc06p03_nba.csv)).

Last column tells if player's career has exceed 5 years or not.

Fill missing values with the field median.

Try to predict if the career has exceeded 5 years or not by using both logistic regression and a support vector machine. Print scores and confusion matrices. Split train/test data as you wish. Compare the results.


````python
import pandas as pd

df = pd.read_csv('data/exrc06p03_nba.csv')

df.head()
````

````python
df.info()
````

````python
df.describe()
````

````python
#filling missing values with median

num_col = df.iloc[:, 1:]

df.iloc[:,1:]=num_col.fillna(num_col.median())

df.head()
````

````python
#logistic regression

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import numpy as np

# the logistic regression model
lr_classifier = LogisticRegression()

X = df.drop('TARGET_5Yrs', axis=1)
y = df['TARGET_5Yrs']


# scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X.select_dtypes(np.number))

````

````python

from sklearn.model_selection import train_test_split

# splitting to train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3,  random_state=42)

lr_classifier.fit(X_train, y_train)
y_pred = lr_classifier.predict(X_test)


print('score (scaled fit):', lr_classifier.score(X_test, y_test))

print(y_pred)
````

````python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
````

````python
#svm

from sklearn.svm import SVC


svm_classifier = SVC()

svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)

print('score (scaled fit):', svm_classifier.score(X_test, y_test))
print(y_pred)
````

````python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
````

The SVM model's accuracy is similar to that of the logistic regression model. However, there is some difference in the confusion matrix.
SVM has slightly better true negatives (80) and the false positives of SVM model (76) were a bit less than for the logistic regression model. 
However, the false negatives increased for SVM (36) and true positives decreased with SVM to 210. So, we cannot say clearly, which model is better.

Try to predict the class (edible or poisonous) from the other fields. Use whatever you want!

Fields are categorial so one-hot-encoding is needed.


````python
import pandas as pd

df = pd.read_csv('data/exrc06p04_mushrooms.csv')
df.head(15)
````

````python
# unique values in each columns
for x in df.columns:
    print(x ,':', len(df[x].unique()))
````

````python
# the example applied from https://machinelearningmastery.com/one-hot-encoding-for-categorical-data/

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder


X = df.drop('class', axis=1)
y = df['class']


# split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# one-hot encode input variables
onehot_encoder = OneHotEncoder()

onehot_encoder.fit(X_train)

X_train = onehot_encoder.transform(X_train)
X_test = onehot_encoder.transform(X_test)

# encode target variable
#label_encoder = LabelEncoder() # this is optional to encode also y

#label_encoder.fit(y_train)

#y_train = label_encoder.transform(y_train)
#y_test = label_encoder.transform(y_test)

print(X_train)

````

````python

# define the model
model = LogisticRegression()

# fit on the training set
model.fit(X_train, y_train)

# predict on test set
y_pred = model.predict(X_test)



print('score:', model.score(X_test, y_test))

print(y_pred)
````

````python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
````

The model has 100% accuracy.

### Problem 5. Loan status
[Here](https://student.labranet.jamk.fi/~varpha/data_analytics/exrc06p05_loan.txt) is some data on loanees. Last column Loan_Status Y/N, should be predicted from the other field. Use whatever you want.  


Do modifications:
* Categorial fields to numeric (2-value fields to 0/1, multivalue as dummies)
* replace missing values with median
* remove rows with outliers: ApplicantIncome, CoapplicantIncome or LoanAmount over 3 standard deviations away from field average


Check what would be model's probability to Loan_status = Yes with values:

```
Gender                   Male
Married                    No
Dependents                  0
Education            Graduate
Self_Employed              No
ApplicantIncome          2400
CoapplicantIncome        2000
LoanAmount                 36
Loan_Amount_Term          360
Credit_History              1
Property_Area           Urban
```

````python
import pandas as pd

df = pd.read_csv('data/loans.txt')

df.set_index('Loan_ID', inplace=True)

df.to_csv("exrc06p05_loans.csv", index = True)

df.head()
````

````python
# dtype of each column 
result = df.dtypes 
 
print(result) 
````

````python
import numpy as np

num_cols = df.select_dtypes(include=np.number).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())

df.head()

````

````python
import pandas as pd
import numpy as np
from scipy import stats

selected_num_col=df[['ApplicantIncome','CoapplicantIncome','LoanAmount']]

df=df[(np.abs(stats.zscore(selected_num_col)) < 3).all(axis=1)]
df.head()
````

````python
cat_col=df.select_dtypes(include=['object'])
cat_col
````

````python
df=df.dropna()
````

````python
cat_col.nunique()
````

We can see that Gender, Married, Education, Self_Employed, and Loan_Status have 2 values.

````python
# unique values in columns
print(df['Gender'].unique())
print(df['Married'].unique())
print(df['Education'].unique())
print(df['Self_Employed'].unique())
print(df['Loan_Status'].unique())

````

````python
df = pd.get_dummies(df, columns=['Property_Area'])
print(df)
````

````python
#Gender, Married, Education, Self_Employed, Credit_History and Loan_Status have 2 values
# replacing values
df['Gender'].replace(['Male', 'Female'],
                        [0, 1], inplace=True)

df['Married'].replace(['No', 'Yes'],
                        [0, 1], inplace=True)
df['Education'].replace(['Graduate', 'Not Graduate'],
                        [0, 1], inplace=True)
df['Self_Employed'].replace(['No', 'Yes'],
                        [0, 1], inplace=True)

df['Loan_Status'].replace(['N', 'Y'],
                        [0, 1], inplace=True)


df.head()
````

````python

from sklearn.preprocessing import StandardScaler
import numpy as np

# scaler
scaler = StandardScaler()


````

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression



X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']

X = scaler.fit_transform(X)


# split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)



# define the model
model = LogisticRegression()

# fit on the training set
model.fit(X_train, y_train)

# predict on test set
y_pred = model.predict(X_test)



print('score:', model.score(X_test, y_test))

print(y_pred)

````

````python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
````
