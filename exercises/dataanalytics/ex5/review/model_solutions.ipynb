## Overfitting

Overfitting is when the model is reading the "noise" (randomness/irregularities) in the specific dataset as a general pattern. The danger of overfitting a model is that you get extremely good model explanatory scores and predictions for the specific training data, but since you have "over-learned" the model it will be significantly worse to predict any additional data fed into the model compared to a more simple model that is not over-fitted. This is common when you attempting to build a very complex model, and especially if you have low amount or low quality training data.

Let's attempt to use the problem 4 code to demonstrate this, one expectation of a overfitted model is that it would perform well on the training set but poorly on the testing set which it has not seen before.

```python
#code re-factored from problem 4

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
import sklearn.metrics

def train_test_r2_printer(df, train_size, test_size):
    df_train, df_test = train_test_split(df, 
                                     train_size = train_size, 
                                     test_size = test_size, 
                                     random_state = 1337)
    
    # rescale the features
    scaler = MinMaxScaler()

    # apply scaler() to all the numeric columns 
    numeric_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
    df_train[numeric_vars] = scaler.fit_transform(df_train[numeric_vars])
    df_train.head()

    # apply rescaling to the test set also
    df_test[numeric_vars] = scaler.fit_transform(df_test[numeric_vars])
    df_test.head()

    # divide into X_train, y_train, X_test, y_test
    y_train = df_train.pop('price')
    X_train = df_train

    y_test = df_test.pop('price')
    X_test = df_test

    # running RFE with number of features=13

    lm = LinearRegression()
    lm.fit(X_train, y_train)

    rfe13 = RFE(lm, n_features_to_select=13)             
    rfe13 = rfe13.fit(X_train, y_train)

    # predict prices of X_test
    y_pred13 = rfe13.predict(X_test)

    # evaluate the model on test set
    r2 = sklearn.metrics.r2_score(y_test, y_pred13)
    print(f"Train size: {train_size}, Test size: {test_size}")
    print("model r2 score:", r2, "\n")
    
train_test_r2_printer(housing, 0.05, 0.95) #this is the one with the smallest train and biggest test
train_test_r2_printer(housing, 0.20, 0.80)
train_test_r2_printer(housing, 0.40, 0.60)
train_test_r2_printer(housing, 0.60, 0.40)
train_test_r2_printer(housing, 0.80, 0.20)
train_test_r2_printer(housing, 0.95, 0.05)
```

By checking the above tests we can see that indeed the model with the highest test size and smallest train size caused the model to violently underperform when exposed to testing data, which as stated is the expected behaviour of an overfitted model.

## K-fold CV (Cross-validation)

K-fold CV is a method that is a more refined version of the regular split of data into two separate training and testing data sets. The purpose of the regular train/test split is to feed the model part of the dataset to train on, and then verify the predictive capabilities with the rest of of the data later. This is another way to prevent over-fitting, since if the accuracy on the test data is much lower than the train data you should consider if the model is being overfitted on the training data.

K-fold cross-validation is then a further refinement of this concept that is ran by iteratively splitting different parts of the dataset into the testing dataset, so that in the end all of the data have been gone over by the model as test data. This reduces potential random inaccuracies that would cause the training data to be significantly different to the testing data.

Let's continue on the same example, but instead train the model using a k-fold cross validation.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
import sklearn.metrics

def train_test_r2_printer_kfold_cv(df, k_folds):
    X = df.drop('price', axis=1)
    y = df['price']
    
    #rescale the features
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)

    #linear regression, rfe using all 13 features
    lm = LinearRegression()
    rfe = RFE(lm, n_features_to_select=13)

    kf = KFold(n_splits=k_folds, shuffle=True, random_state=1337)
    
    r2_scores = cross_val_score(rfe, X_scaled, y, cv=kf, scoring='r2')
    
    print("rfe")
    print("Number of features selected:", rfe.n_features_to_select)
    print(f"Number of splits: {k_folds}")
    print("Average model r2 score:", np.mean(r2_scores))
    print("r2 scores for each split:", r2_scores, "\n")
    
train_test_r2_printer_kfold_cv(housing, 2) #minimiamount of splits
train_test_r2_printer_kfold_cv(housing, 5)
train_test_r2_printer_kfold_cv(housing, 10)
train_test_r2_printer_kfold_cv(housing, 100)
train_test_r2_printer_kfold_cv(housing, 272) #number of samples/2
```
