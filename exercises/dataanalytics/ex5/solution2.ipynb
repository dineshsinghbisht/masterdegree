```python
import pandas as pd
from sklearn.linear_model import LinearRegression

def unknown_data(filename):
    """
    Reads a CSV file, fits a Linear Regression model (no intercept)
    Prints the Correlations and Coefficients
    Returns Correlations and Coefficients
    """
    # Load input csv data which is unknown where first five columns are the features and the last column is the response
    data = pd.read_csv(filename)

    # Separate features and response columns
    X = data.iloc[:, :-1] # Features
    y = data.iloc[:, -1] # Response 

    # Fit linear regression without intercept
    model = LinearRegression(fit_intercept=False)
    model.fit(X, y)

    # Compute correlations for all the features against response
    correlations = X.corrwith(y)

    # Display Linear Regression Coefficients 
    print("**** Print Coefficients ****")
    for i, coef in enumerate(model.coef_, 1):
        print(f"Feature X{i}: {coef:.4f}")

    # Display Featureâ€“Response Correlations
    print("\n**** Print Correlations ****")
    for i, corr in enumerate(correlations, 1):
        print(f"Feature X{i}: {corr:.4f}")

    return correlations, model.coef_


# from getpass import getuser

# # user = getuser()
# # csv_location = f'/home/{user}/dan/private/exrc_04/data/{user}_prob3_sales.csv'
# csv_location = "ah4323_prob3_sales.csv"
# # csv_location =  "FILE_NAME" ### GIVE FILE_NAME TO RUN THE CODE LOCALLY

# # Load input CSV data which contains some (fake) daily sales data
# df = pd.read_csv(csv_location)

# Run the function with unknown data as input file
csv_location = "ah4323_unknown_data.csv"
correlations, coefficients  = unknown_data(csv_location)

# print("\n**** Print Coefficient + Correlation for Features ****")
# for i in range(len(correlations)):
#     coef = coefficients[i]
#     corr = correlations[i]
#     print(f"Feature X{i+1}:  Coefficient = {coef:.4f},  Correlation = {corr:.4f}")
```
