```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function which takes two one-dimensional arrays x and y as parameters. 
# The function should return a tuple (slope, intercept) of the fitted line.
def least_square_fit(x, y):
    """
    Compute the least squares fit line for data points (x, y).

    Parameters: x -> 1D array and y -> 1D array of
    Returns: tuple -> (slope, intercept)
    """

    # Convert input arrays to NumPy arrays to perform arimetic operations easier
    x = np.array(x)
    y = np.array(y)
    
    # Compute means for the input arrays x and y
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    # Calculate slope and intercept using standard formulas
    slope = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
    intercept = y_mean - slope * x_mean
    
    return slope, intercept

# Sample arrays of array to test the function
test_cases = [
    ([1, 2, 3, 4, 5], [2, 4, 5, 4, 5]),
    ([0, 1, 2, 3], [1, 3, 5, 7]),
    ([2, 4, 6, 8], [1, 2, 3, 5]),
    ([8, 10, 15, 20], [5, 7, 9, 10]),
]

# Run tests and print the results
for i, (x, y) in enumerate(test_cases, 1):
    slope, intercept = least_square_fit(x, y)
    print(f"Test case {i}: slope = {slope:.3f}, intercept = {intercept:.3f}")

    plt.figure()
    plt.scatter(x, y, label='Data Points')
    plt.plot(x, slope * np.array(x) + intercept, color='red', label='Fitted Line')
    plt.title(f"Least Squares Fit - Test Case {i}")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
```
