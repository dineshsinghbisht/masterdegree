### handling nan / None
np.na np.notna np.null np.notnull drop_na fill_na

```python
import numpy as np
dict = {
    "Name": ["john", None, np.nan],
    "Age": [26, None, 27]
}

df = pd.DataFrame(dict)
print(df.info())
print(df.isna())
print(df.isna().sum())
print(df.dropna())
print(df)
```
