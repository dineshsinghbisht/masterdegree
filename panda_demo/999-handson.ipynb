### loc vs iloc

```python
### loc vs iloc
import pandas as pd

dict = {
    "Name": ["john", "jack", "gill"],
    "Age": [26, 24, 28],
    "City": ["NY", "CT", "IN"]
}

df = pd.DataFrame(dict)
df.loc[0,"Name"]="Dinesh"
print(df["Name"])
# print(type(df["Name"])) # returns Series (col: Series, col:Series)
# print(df[["Name"]])
# print(type(df[["Name"]])) # returns DataFrame

# print(df.loc[:,"Name"])
# print(df.loc[df["Age"] > 24,"Name"]) # returns Series
# print(df.loc[df["Age"] > 24,["Name"]]) # returns DataFrame

print(df.iloc[:,-1]) # returns Series
print(df.iloc[:,-1:]) # returns Dataframe ie slice preserve the dimension
```

## Sorting



```python
df = pd.DataFrame({"Name": ["John", "David", "Pat", "Tim"],
                    "Age": [30, 27, 28, 50 ]})
print(df)
df.sort_values(by=["Age"],ascending=False, inplace=True)
print("Post Sort\n", df)
```

### Merging

```python
df1 = pd.DataFrame({"CustId": [10, 20],
                    "Name": ["John","Dan"]})
df2 = pd.DataFrame({"CustId": [20, 30],
                    "City": ["Delhi","Goa"]})

print(pd.merge(df1,df2, on ="CustId", how="inner"))
print(pd.merge(df1,df2, on ="CustId", how="outer"))
print(pd.merge(df1,df2, on ="CustId", how="left"))
print(pd.merge(df1,df2, on ="CustId", how="right"))
```

### Concate

```python
df1 = pd.DataFrame({"Name": ["John", "David"],
                    "Age": [26,27]})
df2 = pd.DataFrame({"Name": ["Martin", "Josh"],
                    "Age": [30,40]})

print(pd.concat([df1,df2],axis=0,ignore_index=True))
print(pd.concat([df1,df2],axis=1,ignore_index=True))
```

### Group By

```python
df = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Finance"],
    "Salary": [60000, 65000, 50000, 52000, 70000],
    "EmployeeID": [123, 456, 467, 23423, 9999]
})

print(df.groupby("Department")["Salary"].mean())
print(df.groupby("Department", as_index=False)["Salary"].mean())
print(df.groupby("Department")["Salary"].agg(["mean", "min", "max"]))

df.groupby("Department").agg({
    "Salary": ["mean", "sum"],
    "EmployeeID": "count"
})
```
